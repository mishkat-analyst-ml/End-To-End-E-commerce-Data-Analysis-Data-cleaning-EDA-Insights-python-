```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('Pakistan Largest Ecommerce Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
#Checking for null values
df.isnull().sum()
```

```python
#REMOVING COLUMNS THEY ARE COMPLETELY NULL
df = df.drop(columns=df.columns[df.columns.str.contains('Unnamed')])
#deleting sales_commission_code column 
df = df.drop(columns=['sales_commission_code'])

```

```python
df.columns
```

```python
#FILL NULL VALUES
df['Customer ID']=df['Customer ID'].fillna(0)
df['category_name_1']=df['category_name_1'].fillna('Unknown')

df['FY']=df['FY'].fillna(0)
df[' MV ']=df[' MV '].fillna(0)
df['price']=df['price'].fillna(0)
df['discount_amount']=df['discount_amount'].fillna(0)
df['increment_id']=df['increment_id'].fillna(0)
df['Customer Since']=df['Customer Since'].fillna(0)

df['BI Status']=df['BI Status'].fillna(0)
df['grand_total']=df['grand_total'].fillna(0)
df['qty_ordered']=df['qty_ordered'].fillna(0)
df['sku']=df['sku'].fillna(0)
df['Month']=df['Month'].fillna(0)
df['Year']=df['Year'].fillna(0)
df['Working Date']=df['Working Date'].fillna(0)
df['payment_method']=df['payment_method'].fillna('Unknown')
df.isnull().sum()
```

```python

df['Year'] = pd.to_datetime(df['Year'], format='%Y', errors='coerce')
df['Month'] = pd.to_datetime(df['Month'], format='%m', errors='coerce')
```

```python
df.info()
```

```python
#CREATE REVENUE COLUMN
df['Revenue']=df['price'] * df['qty_ordered']
```

```python
df.head()
```

```python
#AFTER CLEANING THE DATASET I WILL ANALYZE THE DATETSET
#TOTAL REVENUE 
df['Revenue'].sum()
```

```python
#Top products by revenue
df.groupby('sku')['Revenue'].sum().sort_values(ascending=False).head(10)
```

```python
#THEN I WILL FIND SALES BY CATEGORY
df.groupby('category_name_1')['Revenue'].sum().sort_values(ascending=False).head(5).plot(kind='bar')
```

```python

#FINDING YEARLY SALES BY REVENUE
df.groupby('Year')['Revenue'].sum()
```

```python
#THEN I WILL PERFORMING SOME INSIGHTS ON THE DATASET
#TOP 5 PAYMENT METHODS BY REVENUE
df.groupby('payment_method')['Revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#TOP CUSTOMERS
df.groupby('Customer ID')['Revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```

```python
#discount amount by category
df.groupby('category_name_1')['discount_amount'].sum().sort_values(ascending=False).head(5).plot(kind='bar')
```
